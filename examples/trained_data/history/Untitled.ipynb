```python
import pickle
import os
import matplotlib.pyplot as plt
```

```python
model_base = 'PNN'
model_udg = model_base + '_UDG'
dataset = 'alimama'
epoch = 0
```

```python
def pic_his(model, dataset, epoch):
    path = model+'/'+dataset+'/'
    file_list = os.listdir(path)
    file = ''
    for x in file_list:
        if x.split('_')[0] == str(epoch):
            file = x
    pkl_file = open(path + file, 'rb')
    return pickle.load(pkl_file)
```

```python
his_temp = pic_his(model_base, dataset, 1)
```
